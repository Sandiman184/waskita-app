# Klasifikasi menggunakan data primer + word2vec + cosine similarity

```python
# Load dataset

import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Word2vec/Dataset Labelled/Data Primer_Word2vec_Cosine Similarity/data_prepocessing.csv")
df.head(5)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from gensim.models import Word2Vec
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# 1. Load dataset
df = pd.read_csv("/content/drive/MyDrive/Word2vec/Dataset Labelled/Data Primer_Word2vec_Cosine Similarity/data_prepocessing.csv")
df.head(5)

# 2. Initialize LabelEncoder and fit it on the 'Non-Radikal' and 'Radikal' labels
label_encoder = LabelEncoder()
label_encoder.fit(['Non-Radikal', 'Radikal'])  # Fit LabelEncoder with the actual labels 'Non-Radikal' and 'Radikal'

# Preprocessing: stratified train-test split
X = df['normalisasi_kalimat']
y = df['label']

# Convert labels using LabelEncoder to numerical labels
y_encoded = label_encoder.transform(y)  # Encode labels to numeric (0, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42)

# 3. Load pre-trained Word2Vec model
model_word2vec = Word2Vec.load("/content/drive/MyDrive/Word2vec/model_word2vec_cbow/wiki_word2vec_updated.model")
print(f'Word2Vec dimensi: {model_word2vec.vector_size}')
assert model_word2vec.vector_size == 100, f'Expected 100, got {model_word2vec.vector_size}'

# 4. Vectorize text using Word2Vec (average word vectors)
def vectorize_text(text, model_word2vec):
    words = text.split()  # Tokenize based on spaces
    word_vectors = [model_word2vec.wv[word] for word in words if word in model_word2vec.wv.key_to_index]

    return np.mean(word_vectors, axis=0) if word_vectors else np.zeros(model_word2vec.vector_size)

# Apply vectorization to train and test data
X_train_vectors = np.array([vectorize_text(text, model_word2vec) for text in X_train])
X_test_vectors = np.array([vectorize_text(text, model_word2vec) for text in X_test])

# 5. Initialize classifiers
classifiers = {
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}

# 6. Function to display confusion matrix
def plot_confusion_matrix_for_model(model_name, y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix of {model_name} Classification Model')
    plt.show()

# 7. Train and evaluate models
for name, clf in classifiers.items():
    print(f"Training and Evaluating {name}...")

    # Train model
    clf.fit(X_train_vectors, y_train)

    # Make predictions
    y_pred = clf.predict(X_test_vectors)

    # Visualize Confusion Matrix
    plot_confusion_matrix_for_model(name, y_test, y_pred)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {accuracy}")
    print(f"{name} Classification Report:\n{classification_report(y_test, y_pred)}")

    # Cross-validation to ensure reproducibility and generalization
    cv_scores = cross_val_score(clf, X_train_vectors, y_train, cv=5)
    print(f"{name} Cross-Validation Accuracy: {np.mean(cv_scores)}")
    print("-" * 80)
```

Save model ke dengan Joblib

```python
import joblib

# Fungsi untuk menyimpan model klasifikasi
def save_classifier_model(model, model_name):
    file_path = f'/content/drive/MyDrive/Word2vec/models/{model_name}_classifier_model.joblib'
    joblib.dump(model, file_path)
    print(f"Model {model_name} classifier saved at: {file_path}")

# Fungsi untuk menyimpan model Word2Vec
def save_word2vec_model(model_word2vec):
    file_path = '/content/drive/MyDrive/Word2vec/models/word2vec_model.joblib'
    joblib.dump(model_word2vec, file_path)
    print(f"Word2Vec model saved at: {file_path}")

# Fungsi untuk menyimpan LabelEncoder
def save_label_encoder(label_encoder):
    file_path = '/content/drive/MyDrive/Word2vec/models/label_encoder.joblib'
    joblib.dump(label_encoder, file_path)
    print(f"LabelEncoder saved at: {file_path}")
```
